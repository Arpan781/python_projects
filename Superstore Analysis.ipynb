# Import Datasets

```python
filepath = "D:\\Python\\Superstore_USA.xlsx"

try:
    # Read the Excel file using pandas.read_excel()
    ecm = pd.read_excel(filepath)
    print("CSV file read successfully!")

except FileNotFoundError:
    print("Error: File '{}' not found!".format(filepath))
```

# Shipping Mode

```python
#Calculate Category wise total values in Ship Mode
```

```python
ecm["Ship Mode"].value_counts()
```

```python
x= ecm["Ship Mode"].value_counts().index
y= ecm["Ship Mode"].value_counts().values
x,y
```

```python
#plot distribution Of Shiping Mode by category basis In Countplot
```
